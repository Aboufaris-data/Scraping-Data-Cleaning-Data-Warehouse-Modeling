```python
import pandas as pd

df = pd.read_csv(r"C:\Users\user\Desktop\Scraping, Data Cleaning & Data Warehouse Modeling\data\avito_data_raw.csv")

df.info()
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df["surface"] = df["surface"].str.extract(r'(\d+)')[0].astype(float).astype("Int64")
```

```python
df["rooms"] = df["rooms"].str.replace(r"chambres|chambre", "", regex=True).str.strip().astype(int)
```

```python
df["baths"] = df["baths"].str.replace(r"sdbs|sdb", "", regex=True).str.strip().astype(int)
```

```python
df["ville"] = df["location"].str.replace("Appartements dans", "").str.split(",").str[0].str.strip()
```

```python
df["quartier"] =  df["location"].str.replace("Appartements dans", "").str.split(",").str[1].str.strip()
```

```python
df["quartier"].unique()
```

```python
df["title"] = (df["title"].str.encode("ascii", "ignore").str.decode("ascii").str.replace(r"\s+", " ", regex=True).str.strip())
```

```python
df["price"].describe()
```

```python
df["ville"] = df["ville"].str.title().str.strip()
df["quartier"] = df["quartier"].str.title().str.strip()
```

# Feature engenieering


```python
df["price_m"] = df["price"] / df["surface"]
df["price_m"] = df["price_m"].astype(int)
```

```python
def price_category(price):
    if price <= 500000:
        return "Low"
    elif price >= 5000000:
        return "High"
    else:
        return "Medium"

df["price_category"] = df["price"].apply(price_category)
```

```python
df.to_csv(r"C:\Users\user\Desktop\Scraping, Data Cleaning & Data Warehouse Modeling\data\avito_data_clean.csv", index=False)
```
